```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
lam = 1.0
x = np.arange (0, 7, 1)
y1 = stats.poisson.cdf(x, lam)
y2 = stats.poisson.cdf(x, 2.0*lam)
y3 = stats.poisson.cdf(x, 3.0*lam)
plt.scatter(x,y1, label = r" $\lambda_1 = 1$")
plt.scatter(x,y2, marker='x', label = r" $\lambda_2 = 2$")
plt.scatter(x,y3, marker='o', label = r" $\lambda_3 = 3$")
plt.xlabel("n")
plt.ylabel("$P_{кум}(n)$")
plt.title("Вероятность реализации $\leq n$ событий. Кумулятивная функция распределения")
plt.legend()
plt.show()
#plt.savefig('app3_poisson_1.jpg')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
lam = 1.0
x = np.arange (0, 7, 1)
y1 = stats.poisson.pmf(x, lam)
y2 = stats.poisson.pmf(x, 2.0*lam)
y3 = stats.poisson.pmf(x, 3.0*lam)
plt.scatter(x,y1, label = r" $\lambda_1 = 1$")
plt.scatter(x,y2, marker='x', label = r" $\lambda_2 = 2$")
plt.scatter(x,y3, marker='o', label = r" $\lambda_3 = 3$")
plt.legend()
plt.show()
#plt.savefig('app3_poisson_1.jpg')
```
