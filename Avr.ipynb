Среднее число неисправных (находящихся на обслуживании) элементов:
$  N_T = 1 \pi_1 + 2 \pi_2 = \frac{2\rho}{1+ 2\rho + 2\rho^2} (1 + 2\rho).  
$  

```python
import matplotlib.pyplot as plt
import numpy as np
ro = np.arange (0.01, 2.0, 0.01)
NT = (2*ro*(1+2*ro))/(1 + 2*ro + 2*ro**2)
plt.plot(ro,NT, label = r" $N_T$")
plt.title('Расчет среднего числа элементов в ремонте')  
plt.legend()
plt.grid(True)
plt.show()
```
